```python
class Node:

    def __init__(self,value):
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:

    def __init__(self,value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self,msg = "List Created"):
        print('---------------')
        print(msg)
        print('---------------')
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
        print('---------------')
        print(f"Head = {self.head.value}")
        print(f"Tail = {self.tail.value}")
        print(f"Length of the List = {self.length}")
        print('---------------')
```

### Code

```python
import graphviz

class Node:

  def __init__(self, value):
    self.value = value
    self.next = None
    self.prev = None

class DoublyLinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):

        dot = graphviz.Digraph()
        dot.attr(rankdir='LR')  # Change direction to horizontal
        temp = self.head
        index = 0

        while temp is not None:

            label = f"{temp.value} (index={index})"

            if self.length == 1:
                dot.node(str(temp.value), label, fillcolor="cyan", style="filled")
                break
            else:
                if temp == self.head:
                    dot.node(str(temp.value), label, fillcolor="chartreuse", style="filled")
                elif temp == self.tail:
                    dot.node(str(temp.value), label, fillcolor="lightpink", style="filled")
                else:
                    dot.node(str(temp.value), label)

            if temp.next:
                dot.edge(str(temp.value), str(temp.next.value))

            if temp.prev:
                dot.edge(str(temp.value), str(temp.prev.value))

            temp = temp.next
            index += 1

        dot.edge("Start", str(self.head.value), dir="back")
        dot.edge(str(self.tail.value), "End")

        return dot

    def append(self,value):
        # create the new node
        new_node = Node(value)

        if self.length == 0:
            self.tail = new_node
            self.head = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

        self.length += 1
```

```python
# Example usage
my_list = DoublyLinkedList(7)
my_list.append(11)
my_list.append(12)
my_list.append(13)
display(my_list.print_list())
```

```python
import graphviz

class Node:

  def __init__(self, value):
    self.value = value
    self.next = None
    self.prev = None

class DoublyLinkedList:

    def __init__(self, value):

        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def make_empty(self):

        self.head = None
        self.tail = None
        self.length = 0

    def print_list(self,msg = "List Created"):

        print("\n")
        print("-----------")
        print(msg)
        print("-----------")
        print("\n")

        if self.length == 0:
            return "Nothing to display"


        dot = graphviz.Digraph()
        dot.attr(rankdir='LR')  # Change direction to horizontal
        temp = self.head
        index = 0

        while temp is not None:
            label = f"{temp.value} (index={index})"

            if self.length == 1:
                dot.node(str(temp.value), label, fillcolor="cyan", style="filled")
                break
            else:
                if temp == self.head:
                    dot.node(str(temp.value), label, fillcolor="chartreuse", style="filled")
                elif temp == self.tail:
                    dot.node(str(temp.value), label, fillcolor="lightpink", style="filled")
                else:
                    dot.node(str(temp.value), label)

            if temp.next:
                dot.edge(str(temp.value), str(temp.next.value))

            if temp.prev:
                dot.edge(str(temp.value), str(temp.prev.value))

            temp = temp.next
            index += 1

        dot.edge("Start", str(self.head.value), dir="back")
        dot.edge(str(self.tail.value), "End")

        return dot

    def append(self,value):

        # create the new node
        new_node = Node(value)

        if self.length == 0:
            self.tail = new_node
            self.head = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node

        self.length += 1

    def pop(self):

        if self.length == 0:
            return None

        temp = self.tail

        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            # disconnect the last node
            self.tail = self.tail.prev
            self.tail.next = None

        temp.prev = None
        self.length -= 1

        return temp

    def prepend(self,value):

        new_node = Node(value)

        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            temp = self.head
            new_node.next = temp
            self.head = new_node
            temp.prev = new_node

        self.length += 1
        return True

    def pop_first(self):

        if self.length == 0:
            return None

        temp = self.head

        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = temp.next
            self.head.prev = None

        temp.next = None
        self.length -= 1
        return temp
```

```python
# -----------------------
# Test case : pop_first() | with no elements
# -----------------------
my_doubly_linked_list = DoublyLinkedList(1)
my_doubly_linked_list.make_empty()
display(my_doubly_linked_list.print_list("pop_first() | with no elements"))

ret_node = my_doubly_linked_list.pop_first()

if ret_node:
    print("\n")
    print(f"Popped node = {ret_node.value}")
display(my_doubly_linked_list.print_list("List Updated"))

# -----------------------
# Test case : pop_first() | with 1 element
# -----------------------
my_doubly_linked_list = DoublyLinkedList(1)
display(my_doubly_linked_list.print_list("pop_first() | with 1 element"))

ret_node = my_doubly_linked_list.pop_first()

if ret_node:
    print("\n")
    print(f"Popped node = {ret_node.value}")
display(my_doubly_linked_list.print_list("List Updated"))

# -----------------------
# Test case : pop_first() | with multiple elements
# -----------------------
my_doubly_linked_list = DoublyLinkedList(1)
my_doubly_linked_list.append(2)
my_doubly_linked_list.append(3)
my_doubly_linked_list.append(4)
my_doubly_linked_list.append(5)
display(my_doubly_linked_list.print_list("pop_first() | with multiple elements"))

ret_node = my_doubly_linked_list.pop_first()

if ret_node:
    print("\n")
    print(f"Popped node = {ret_node.value}")
display(my_doubly_linked_list.print_list("List Updated"))
```
